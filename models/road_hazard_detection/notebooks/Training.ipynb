```python
import torch
import numpy as np
import random

torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
```

```python
from ultralytics import YOLO

# Load pretrained YOLOv8m (medium size for balance of speed/accuracy)
model = YOLO('yolov8m.pt')
```

```python
model.train(
    data="/content/drive/MyDrive/RoadHazardDetection/datasets/data.yaml",
    epochs=40,          # number of training epochs
    imgsz=640,          # image size
    batch=16,           # batch size
    lr0=0.01,           # learning rate
    optimizer='SGD',    # optimizer
    momentum=0.937,
    weight_decay=5e-4,
    project="RoadHazardDetection",  # folder for outputs
    name="exp_yolov8m_base",          # run name
    augment=True,       # data augmentation
    val=True            # perform validation
)
```
