```python
from ultralytics import YOLO

# Load pretrained YOLOv8 medium model
model = YOLO('yolov8m.pt')

# Train the model
model.train(
    data="/content/drive/MyDrive/road-hazard-detection/datasets/YOLOv8_extracted/data.yaml",  # Path to dataset YAML
    epochs=50,              # Number of training epochs
    imgsz=640,              # Image size
    batch=16,               # Batch size
    lr0=0.01,               # Initial learning rate
    optimizer='SGD',        # Optimizer
    momentum=0.937,         # SGD momentum
    weight_decay=5e-4,      # Weight decay (L2 regularization)
    project="road-hazard-detection",  # W&B / save project folder
    name="exp_yolov8m_base2",         # Subfolder name for this run
    augment=True,           # Apply data augmentation
    verbose=True,           # Display training progress
)
```

```python
metrics = model.val()
print(metrics)
```

```python
import os
import shutil

val_images_nested = "/content/drive/MyDrive/road-hazard-detection/datasets/YOLOv8_extracted/valid/images/images"
val_images_correct = "/content/drive/MyDrive/road-hazard-detection/datasets/YOLOv8_extracted/valid/images"

# Create correct folder if missing
os.makedirs(val_images_correct, exist_ok=True)

# Move all images up
for file in os.listdir(val_images_nested):
    if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp')):
        shutil.move(os.path.join(val_images_nested, file), os.path.join(val_images_correct, file))

# Remove the empty nested folder
shutil.rmtree(val_images_nested)

print("✅ Validation images flattened successfully!")
```

```python
import pandas as pd
from ultralytics import YOLO

# Load your trained model
model = YOLO("/content/road-hazard-detection/exp_yolov8m_base2/weights/best.pt")

# Run evaluation on validation dataset
metrics = model.val(
    data="/content/drive/MyDrive/road-hazard-detection/datasets/YOLOv8_extracted/data.yaml",
    save=False,
    verbose=True
)

# Extract metrics from metrics.box
results_dict = {
    "epoch": [0],  # Single evaluation
    "precision": [metrics.box.mp],
    "recall": [metrics.box.mr],
    "mAP50": [metrics.box.map50],
    "mAP50-95": [metrics.box.map]
}

# Save as CSV
df = pd.DataFrame(results_dict)
df.to_csv("/content/drive/MyDrive/road-hazard-detection/models/results/results.csv", index=False)

print("✅ results.csv created successfully")
```
